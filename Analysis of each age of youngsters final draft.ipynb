```python
import pandas as pd

age = 16
lived_to_expectations = 'lived to expectations'
exceeded_expectations = 'exceeded their potential'
flopped = 'did not hit their potential'
still_young = 'still has time'

list_of_analysis_for_every_fifa = []
for fifa, x in zip(range(10,22),range(0,12), ):
    which_fifa = {}
    all_fifas = [f'Fifa {i} overall_rating' for i in range(fifa, 23)]

    original_database = pd.read_csv(f'cleaned_fifa_{fifa}_top300_potential_players.csv')

    original_database['Best Overall Rating'] = list(map(lambda x: max(x),original_database[all_fifas].values))
    columns_of_interest = original_database[['Player name', 'Known as','nationality','team', 'position',\
                                                   f'Fifa {fifa} Potential rating', f'Fifa {fifa} overall_rating',\
                                                    f'Fifa {fifa} age', 'Best Overall Rating', \
                                                    f'Fifa {fifa} outcome', 'Year potential reached']]

    total_players_analyzed = len(original_database)

    age_data = columns_of_interest.groupby([f'Fifa {fifa} age',f'Fifa {fifa} outcome']).describe()[f'Fifa {fifa} Potential rating']
    list_of_analysis_of_each_age_in_each_fifa = []
    for age, a in zip(range(16,24),range(0,8)):

        analysis_of_each_age_in_each_fifa = {}

        analysis_of_each_age_in_each_fifa['Which Age?'] = age
        
        try:
            analysis_of_each_age_in_each_fifa[f'how_many_{age}'] = age_data.loc[age]['count'].sum()
        except KeyError:
            analysis_of_each_age_in_each_fifa[f'how_many_{age}'] = 0
            
        if analysis_of_each_age_in_each_fifa[f'how_many_{age}'] == 0:
             analysis_of_each_age_in_each_fifa[f'prop_how_many_{age}'] = 0 
        else:
            analysis_of_each_age_in_each_fifa[f'prop_how_many_{age}'] = age_data.loc[age]['count'].sum() /total_players_analyzed

        try:
            analysis_of_each_age_in_each_fifa[f'how_many_{age}_lived_to_expectations'] = age_data.loc[age,lived_to_expectations]['count']
            analysis_of_each_age_in_each_fifa[f'prop_how_many_{age}_lived_to_expectations'] = analysis_of_each_age_in_each_fifa[f'how_many_{age}_lived_to_expectations'] / analysis_of_each_age_in_each_fifa[f'how_many_{age}']
        except (KeyError,ZeroDivisionError):
            analysis_of_each_age_in_each_fifa[f'how_many_{age}_lived_to_expectations'] = 0
            analysis_of_each_age_in_each_fifa[f'prop_how_many_{age}_lived_to_expectations'] = 0

        try:
            analysis_of_each_age_in_each_fifa[f'how_many_{age}_exceeded_expectations'] = age_data.loc[age,exceeded_expectations]['count']
            analysis_of_each_age_in_each_fifa[f'prop_how_many_{age}_exceeded_expectations'] = analysis_of_each_age_in_each_fifa[f'how_many_{age}_exceeded_expectations'] / analysis_of_each_age_in_each_fifa[f'how_many_{age}']
        except (KeyError,ZeroDivisionError):
            analysis_of_each_age_in_each_fifa[f'how_many_{age}_exceeded_expectations'] = 0 
            analysis_of_each_age_in_each_fifa[f'prop_how_many_{age}_exceeded_expectations'] = 0

        analysis_of_each_age_in_each_fifa[f'how_many_{age}_succeed'] = analysis_of_each_age_in_each_fifa[f'how_many_{age}_lived_to_expectations'] + analysis_of_each_age_in_each_fifa[f'how_many_{age}_exceeded_expectations']
        
        if analysis_of_each_age_in_each_fifa[f'how_many_{age}_succeed'] == 0:
            analysis_of_each_age_in_each_fifa[f'prop_how_many_{age}_succeed'] = 0
        else:
            analysis_of_each_age_in_each_fifa[f'prop_how_many_{age}_succeed'] = analysis_of_each_age_in_each_fifa[f'how_many_{age}_succeed'] / analysis_of_each_age_in_each_fifa[f'how_many_{age}']

        try:
            analysis_of_each_age_in_each_fifa[f'how_many_{age}_flopped'] = age_data.loc[age,flopped]['count']
            analysis_of_each_age_in_each_fifa[f'prop_how_many_{age}_flopped'] = analysis_of_each_age_in_each_fifa[f'how_many_{age}_flopped'] / analysis_of_each_age_in_each_fifa[f'how_many_{age}']
        except (KeyError,ZeroDivisionError):
            analysis_of_each_age_in_each_fifa[f'how_many_{age}_flopped'] = 0 
            analysis_of_each_age_in_each_fifa[f'prop_how_many_{age}_flopped'] = 0 

        try:
            analysis_of_each_age_in_each_fifa[f'how_many_{age}_youngsters'] = age_data.loc[age,still_young]['count']
            analysis_of_each_age_in_each_fifa[f'prop_how_many_{age}_youngsters'] = analysis_of_each_age_in_each_fifa[f'how_many_{age}_youngsters'] / analysis_of_each_age_in_each_fifa[f'how_many_{age}']
        except (KeyError,ZeroDivisionError):
            analysis_of_each_age_in_each_fifa[f'how_many_{age}_youngsters'] = 0 
            analysis_of_each_age_in_each_fifa[f'prop_how_many_{age}_youngsters'] = 0
            
        list_of_analysis_of_each_age_in_each_fifa.append(analysis_of_each_age_in_each_fifa)
        
    which_fifa['Which Fifa?'] = fifa
    which_fifa['total_players_analyzed'] = total_players_analyzed
    which_fifa['analysis_of_each_age_in_each_fifa'] = list_of_analysis_of_each_age_in_each_fifa
    list_of_analysis_for_every_fifa.append(which_fifa)

list_of_analysis_for_every_fifa[2]
```

```python
list_of_youngster_analysis = []

for age, age_index in zip(range(16,24), range(0,8)):
    each_age = {}
    total_players_analyzed = sum((list_of_analysis_for_every_fifa[i]['total_players_analyzed'] for i in range(0,12)))

    how_many_ = sum((list_of_analysis_for_every_fifa[i]['analysis_of_each_age_in_each_fifa'][age_index][f'how_many_{age}'] for i in range(0,12)))
    prop_how_many_ = how_many_ / total_players_analyzed 

    how_many_lived_to_expectations =  sum((list_of_analysis_for_every_fifa[i]['analysis_of_each_age_in_each_fifa'][age_index][f'how_many_{age}_lived_to_expectations'] for i in range(0,12)))
    prop_how_many_lived_to_expectations = how_many_lived_to_expectations/how_many_

    how_many_exceeded_expectations = sum((list_of_analysis_for_every_fifa[i]['analysis_of_each_age_in_each_fifa'][age_index][f'how_many_{age}_exceeded_expectations'] for i in range(0,12)))
    prop_how_many_exceeded_expectations = how_many_exceeded_expectations /how_many_

    how_many_succeed = sum((list_of_analysis_for_every_fifa[i]['analysis_of_each_age_in_each_fifa'][age_index][f'how_many_{age}_succeed'] for i in range(0,12)))
    prop_how_many_succeed = how_many_succeed/ how_many_

    how_many_flopped = sum((list_of_analysis_for_every_fifa[i]['analysis_of_each_age_in_each_fifa'][age_index][f'how_many_{age}_flopped'] for i in range(0,12)))
    prop_how_many_flopped = how_many_flopped /how_many_

    how_many_youngsters = sum((list_of_analysis_for_every_fifa[i]['analysis_of_each_age_in_each_fifa'][age_index][f'how_many_{age}_youngsters'] for i in range(0,12)))
    prop_how_many_youngsters = how_many_youngsters/how_many_


    each_age['total_players_analyzed'] = total_players_analyzed
    each_age[f'how_many_{age}_hit_prime'] = how_many_ - how_many_youngsters
    each_age[f'how_many_total_{age}'] = how_many_
    each_age[f'prop_how_many_total_{age}'] = prop_how_many_
    each_age[f'how_many_{age}_lived_to_expectations'] = how_many_lived_to_expectations
    each_age[f'prop_how_many_{age}_lived_to_expectations'] = prop_how_many_lived_to_expectations
    each_age[f'how_many_{age}_exceeded_expectations'] = how_many_exceeded_expectations
    each_age[f'prop_how_many_{age}_exceeded_expectations'] = prop_how_many_exceeded_expectations
    each_age[f'how_many_{age}_succeed'] = how_many_succeed
    each_age[f'prop_how_many_{age}_succeed'] = prop_how_many_succeed
    each_age[f'how_many_{age}_flopped'] = how_many_flopped
    each_age[f'prop_how_many_{age}_flopped'] = prop_how_many_flopped
    each_age[f'how_many_{age}_youngsters'] = how_many_youngsters
    each_age[f'prop_how_many_{age}_youngsters'] = prop_how_many_youngsters

    list_of_youngster_analysis.append(each_age)



list_of_youngster_analysis[7]
```
