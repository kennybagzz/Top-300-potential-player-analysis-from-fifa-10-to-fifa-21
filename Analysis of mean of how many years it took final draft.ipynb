
# Analysis of each Fifa

```python
import pandas as pd
import statistics as stats
import numpy as np
import pickle

#define player outcomes
lived_to_expectations = 'lived to expectations'
exceeded_expectations = 'exceeded their potential'
flopped = 'did not hit their potential'
still_young = 'still has time'

#create empty list to hold each analysis
list_of_analysis_for_every_fifa = []

#iterate through every fifa
for each_fifa in range(10,22):

    #create empty dictionary to put all analysis in.
    fifa_insight = {}
    all_fifas = [f'Fifa {i} overall_rating' for i in range(each_fifa, 23)]
    
    #create database of interest
    original_database = pd.read_csv(f'cleaned_fifa_{each_fifa}_top300_potential_players.csv')
    original_database['Best Overall Rating'] = list(map(lambda x: max(x),original_database[all_fifas].values))
    columns_of_interest = original_database[['Player name', 'Known as','nationality','team', 'position',\
                                               f'Fifa {each_fifa} Potential rating', f'Fifa {each_fifa} overall_rating',\
                                             'Best Overall Rating',f'Fifa {each_fifa} outcome', 'Year potential reached']]
    
    #define how each outcome is reached
    Players_who_succeeded = columns_of_interest[columns_of_interest['Year potential reached'] >= 1]
    Players_who_exceeded_expectations = columns_of_interest[columns_of_interest[f'Fifa {each_fifa} outcome']\
                                                              == exceeded_expectations]
    Players_who_lived_to_expectations = columns_of_interest[columns_of_interest[f'Fifa {each_fifa} outcome']\
                                                              == lived_to_expectations]
    players_who_flopped_dataframe = original_database[columns_of_interest[f'Fifa {each_fifa} outcome']\
                                                              == flopped]
    players_who_are_still_young = original_database[columns_of_interest[f'Fifa {each_fifa} outcome']\
                                                              == still_young]
    
    #how many people in dataframe 
    total_players_analyzed = len(original_database)

    #how many people hit potential or exceeded potential
    how_many_succeeded = len(Players_who_succeeded)
    #proportion of how many people hit potential or exceeded potential
    prop_of_how_many_succeeded = how_many_succeeded/total_players_analyzed

    #how many lived to expectations 
    how_many_lived_to_expectations = len(Players_who_lived_to_expectations)
    #proportion of how many lived to expectations 
    prop_of_how_many_lived_to_expectations = how_many_lived_to_expectations / total_players_analyzed

    #how many exceeded expectations
    how_many_exceeded_expectations = len(Players_who_exceeded_expectations)
    #proportion of how many exceeded expectations
    prop_of_how_many_exceeded_expectations = how_many_exceeded_expectations / total_players_analyzed

    #how many flopped 
    players_who_flopped = len(players_who_flopped_dataframe)
    #proportion of how many flopped
    prop_of_players_who_flopped = players_who_flopped / total_players_analyzed

    #how many haven't hit their prime
    players_still_young = len(players_who_are_still_young)
    #prop of players who are still young
    prop_of_players_still_young = players_still_young/ total_players_analyzed

    #mean of how many years it took to hit potential
    mean_of_dataset = Players_who_succeeded['Year potential reached'].mean()

    #25th, 50th, 75th percentile of how many years it took to hit potential
    twenty_fifth_percentile = Players_who_succeeded['Year potential reached'].quantile(0.25)
    fiftieth_percentile = Players_who_succeeded['Year potential reached'].quantile(0.5)
    seventy_fifth_percentile = Players_who_succeeded['Year potential reached'].quantile(0.75)

    #standard deviation of how many years it took to hit potential
    std = Players_who_succeeded['Year potential reached'].std()

    #min and max of how many years it took to hit potential
    min_num_years = min(Players_who_succeeded['Year potential reached'])
    max_num_years = max(Players_who_succeeded['Year potential reached'])
    
    #store all analysis into a dictionary
    fifa_insight['Which Fifa?'] = f'Fifa {each_fifa}'
    fifa_insight['original_database'] = original_database
    fifa_insight['Players_who_succeeded_database'] = Players_who_succeeded
    fifa_insight['list of all years potential reached'] = list(Players_who_succeeded['Year potential reached'])
    fifa_insight['total_players_analyzed'] = total_players_analyzed
    fifa_insight['how_many_succeeded'] = how_many_succeeded
    fifa_insight['prop_of_how_many_succeeded'] = prop_of_how_many_succeeded
    fifa_insight['how_many_lived_to_expectations'] = how_many_lived_to_expectations
    fifa_insight['prop_of_how_many_lived_to_expectations'] = prop_of_how_many_lived_to_expectations
    fifa_insight['how_many_exceeded_expectations'] = how_many_exceeded_expectations
    fifa_insight['prop_of_how_many_exceeded_expectations'] = prop_of_how_many_exceeded_expectations
    fifa_insight['players_who_flopped'] = players_who_flopped 
    fifa_insight['prop_of_players_who_flopped'] = prop_of_players_who_flopped
    fifa_insight['players_still_young'] = players_still_young
    fifa_insight['prop_of_players_still_young'] = prop_of_players_still_young
    fifa_insight['mean_of_dataset'] = mean_of_dataset
    fifa_insight['twenty_fifth_percentile'] = twenty_fifth_percentile
    fifa_insight['fiftieth_percentile'] = fiftieth_percentile
    fifa_insight['seventy_fifth_percentile'] = seventy_fifth_percentile
    fifa_insight['standard deviation'] = std
    fifa_insight['min_num_years'] = min_num_years
    fifa_insight['max_num_years'] = max_num_years
    
    #append dictionary to list 
    list_of_analysis_for_every_fifa.append(fifa_insight)

list_of_analysis_for_every_fifa
```

# Analysis of all Fifas

```python
analysis_of_all_fifas = {}

total_players_analyzed = sum([list(list_of_analysis_for_every_fifa[i].values())[4] for i in range(0,12)])

players_still_young = sum([list(list_of_analysis_for_every_fifa[i].values())[13] for i in range(0,12)])
prop_of_players_still_young = players_still_young / total_players_analyzed

players_who_arent_youngsters = total_players_analyzed - players_still_young

how_many_succeeded = sum([list(list_of_analysis_for_every_fifa[i].values())[5] for i in range(0,12)])
prop_of_how_many_succeeded = how_many_succeeded / total_players_analyzed
prop_of_how_many_succeeded_wo_youngsters = how_many_succeeded / players_who_arent_youngsters

how_many_lived_to_expectations = sum([list(list_of_analysis_for_every_fifa[i].values())[7] for i in range(0,12)])
prop_of_how_many_lived_to_expectations = how_many_lived_to_expectations /total_players_analyzed
prop_of_how_many_lived_to_expectations_wo_youngsters = how_many_lived_to_expectations / players_who_arent_youngsters

how_many_exceeded_expectations = sum([list(list_of_analysis_for_every_fifa[i].values())[9] for i in range(0,12)])
prop_of_how_many_exceeded_expectations= how_many_exceeded_expectations / total_players_analyzed
prop_of_how_many_exceeded_wo_youngsters =  how_many_exceeded_expectations / players_who_arent_youngsters

how_many_flopped = sum([list(list_of_analysis_for_every_fifa[i].values())[11] for i in range(0,12)])
prop_of_how_many_flopped = how_many_flopped / total_players_analyzed
prop_of_how_many_flopped_wo_youngsters = how_many_flopped / players_who_arent_youngsters

overall_mean = sum([sum(list_of_analysis_for_every_fifa[i]['Players_who_succeeded_database']\
                        ['Year potential reached']) for i in range(0,12)]) /\
                        (total_players_analyzed-players_still_young - how_many_flopped)

all_list_years_potential_reached = []
for i in range(0,12):
    all_list_years_potential_reached.extend(list(list_of_analysis_for_every_fifa[i].values())[3])

series_of_all_years_potential_reached = pd.Series(all_list_years_potential_reached)
std_of_all_years_potential_reached = series_of_all_years_potential_reached.std()

#25th, 50th, 75th percentile of how many years it took to hit potential
twenty_fifth_percentile = series_of_all_years_potential_reached.quantile(0.25)
fiftieth_percentile = series_of_all_years_potential_reached.quantile(0.5)
seventy_fifth_percentile = series_of_all_years_potential_reached .quantile(0.75)

#min and max of how many years it took to hit potential
min_num_years = min(series_of_all_years_potential_reached)
max_num_years = max(series_of_all_years_potential_reached)

analysis_of_all_fifas['all_list_years_potential_reached'] = all_list_years_potential_reached
analysis_of_all_fifas['total_players_analyzed'] = total_players_analyzed
analysis_of_all_fifas['players_still_young'] = players_still_young
analysis_of_all_fifas['prop_of_players_still_young'] = prop_of_players_still_young
analysis_of_all_fifas['players_who_arent_youngsters'] = players_who_arent_youngsters
analysis_of_all_fifas['how_many_succeeded'] = how_many_succeeded
analysis_of_all_fifas['prop_of_how_many_succeeded'] = prop_of_how_many_succeeded
analysis_of_all_fifas['prop_of_how_many_succeeded_wo_youngsters'] = prop_of_how_many_succeeded_wo_youngsters
analysis_of_all_fifas['how_many_lived_to_expectations'] = how_many_lived_to_expectations
analysis_of_all_fifas['prop_of_how_many_lived_to_expectations'] = prop_of_how_many_lived_to_expectations
analysis_of_all_fifas['prop_of_how_many_lived_to_expectations_wo_youngsters'] = prop_of_how_many_lived_to_expectations_wo_youngsters
analysis_of_all_fifas['how_many_exceeded_expectations'] = how_many_exceeded_expectations
analysis_of_all_fifas['prop_of_how_many_exceeded_expectations'] = prop_of_how_many_exceeded_expectations
analysis_of_all_fifas['prop_of_how_many_exceeded_wo_youngsters'] = prop_of_how_many_exceeded_wo_youngsters
analysis_of_all_fifas['how_many_flopped'] = how_many_flopped
analysis_of_all_fifas['prop_of_how_many_flopped'] = prop_of_how_many_flopped
analysis_of_all_fifas['prop_of_how_many_flopped_wo_youngsters'] = prop_of_how_many_flopped_wo_youngsters
analysis_of_all_fifas['overall_mean_years_potential_reached'] = overall_mean
analysis_of_all_fifas['all_list_years_potential_reached']
analysis_of_all_fifas['standard_deviation_of_all_years_potential_reached'] = std_of_all_years_potential_reached
analysis_of_all_fifas['twenty_fifth_percentile'] = twenty_fifth_percentile
analysis_of_all_fifas['fiftieth_percentile'] = fiftieth_percentile
analysis_of_all_fifas['seventy_fifth_percentile'] = seventy_fifth_percentile
analysis_of_all_fifas['min_num_years'] = min_num_years
analysis_of_all_fifas['max_num_years'] = max_num_years

analysis_of_all_fifas
```
